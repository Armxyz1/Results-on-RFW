```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
races = ['African','Asian','Caucasian','Indian']
models = ['AdaFace','ArcFace','ElasticFace','GhostFace','SphereFace']
```

```python
for model in models:
    fig = plt.figure()
    for race in races:
            with open(f'{model}/roc/{race}_roc.npy','rb') as f:
                fpr  = np.load(f)
                tpr  = np.load(f)
                plt.plot(fpr,tpr,label = f"{race}")
    plt.legend()
    plt.title(model)
    plt.xlabel('FPR')
    plt.ylabel('TPR')
    plt.xscale('log')
    plt.savefig(f'./Graphs/Comb/{model}_roc.png')
```
